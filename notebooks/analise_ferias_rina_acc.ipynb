```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r'..\data\ferias.xlsx', sheet_name='Planilha1')
```

```python
df.head()
```

```python
new_header = df.iloc[3]
```

```python
df = df.iloc[4:]
```

```python
df.head(10)
```

```python
df.columns = new_header
```

```python
df.head()
```

```python
df = df.reset_index(drop=True)
```

```python
df.columns
```

```python
df.columns=df.columns.to_list()
```

```python
df.head()
```

```python
df = df[['Nome','Cargo','Data Início','Fim']]
```

```python
df= df.rename(columns={'Fim':'Data fim'})
```

```python
df.head()
```

```python
df.info()
```

```python
df[df['Cargo'].isna() | df['Data Início'].isna()]
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['Data Início'] = pd.to_datetime(df['Data Início'])
```

```python
df.info()
```

```python
df['Data fim'] = pd.to_datetime(df['Data fim'])
```

```python
df['mes'] = df['Data Início'].dt.month_name()
```

```python
df['numero mes'] = df['Data Início'].dt.month
```

```python
df.head(10)
```

```python
df['ano'] = df['Data Início'].dt.year
```

```python
df.info()
```

```python
df_2025 = df[df['ano']== 2025]
```

```python
df_2025.head(10)
```

```python
df_by_mes = df_2025.groupby(['mes','numero mes'] ,as_index=False)['Nome'].count().sort_values(by='numero mes',ascending=True)
```

```python
df_by_mes = df_by_mes.drop(columns='numero mes',axis=1)
```

```python
df_by_mes.head()
```

```python
plt.figure(figsize=(12,4))
plt.title('Quantidade de funcionarios de férias por mês 2025')
bars = sns.barplot(df_by_mes,x='mes',y='Nome',legend='auto')
for bar in bars.patches:
    height = bar.get_height()
    bars.text(bar.get_x() + bar.get_width()/2,height,height,ha='center',va='bottom' )
plt.ylabel('Quantidade de funcionários')
plt.xlabel('Mês')
plt.show()
```
